```python
optCalls = df.iloc[:, [2, 4, 7, 8, 13, 15, 17, 18, 31]]
optCalls
```

```python
optCalls[' [QUOTE_DATE]'] = pd.to_datetime(optCalls[' [QUOTE_DATE]'])
```

```python
optCalls
```

```python
optCalls = optCalls[(optCalls[' [DTE]'] >= 7) & (optCalls[' [DTE]'] <= 14)]
optCalls
```

```python
#filtered_df = optCalls.groupby(' [QUOTE_DATE]')[' [STRIKE_DISTANCE]'].min().reset_index()

min_indexes = optCalls.groupby(' [QUOTE_DATE]')[' [STRIKE_DISTANCE]'].idxmin()

filtered_df = optCalls.loc[min_indexes]
```

```python
filtered_df
```

Hacer lo mismo con la ayuda de chat gpt, pero esta vez con valores más cercanos a 0.5 de cada fecha. Ojo con DTE, la idea es que DTE solo corresponda a 7-14 o 21.

```python
#basicamente, se calcula la diferencia de delta absoluta, respecto a 0.5, y luego toma el mínimo

idx_min_abs_diff = optCalls.groupby(' [QUOTE_DATE]')[' [C_DELTA]'].apply(lambda x: (x - 0.5).abs().idxmin())

filtered_df_2 = optCalls.loc[idx_min_abs_diff]
```

```python
filtered_df_2
```

Al parecer "filtered_df_2" es más preciso con los valores de delta.

```python
filtered_df_2.info()
```

Graficar

```python
# Crear la figura y los dos ejes (izquierdo y derecho)
fig, ax1 = plt.subplots(figsize=(12, 6))

# Graficar los valores de 'UNDERLYING_LAST' en el eje izquierdo
ax1.plot(filtered_df_2[' [QUOTE_DATE]'], filtered_df_2[' [UNDERLYING_LAST]'], color='b', label='UNDERLYING_LAST')
ax1.set_ylabel('UNDERLYING_LAST', color='b')
ax1.tick_params('y', colors='b')

# Crear el eje derecho y graficar los valores de 'C_LAST' en el eje derecho
ax2 = ax1.twinx()
ax2.plot(filtered_df_2[' [QUOTE_DATE]'], filtered_df_2[' [C_LAST]'], color='r', label='C_LAST')
ax2.set_ylabel('C_LAST', color='r')
ax2.tick_params('y', colors='r')

# Configurar el eje x con las fechas
plt.xticks(rotation=45)
plt.xlabel(' [QUOTE_DATE]')

# Agregar leyendas y título
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')
plt.title('Valores de UNDERLYING_LAST y C_LAST por fecha')

#cuadrículas
#ax1.grid(True)
#ax2.grid(True)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
