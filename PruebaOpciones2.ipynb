```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Leer los tres archivos de texto y almacenarlos en dataframes
df1 = pd.read_csv('tsla_eod_202201.txt', delimiter=',')  # Ajusta el delimitador según el formato del archivo
df2 = pd.read_csv('tsla_eod_202202.txt', delimiter=',')
df3 = pd.read_csv('tsla_eod_202203.txt', delimiter=',')
```

```python
df_concatenado = pd.concat([df1, df2, df3])

# Guardar el nuevo dataframe en un archivo de texto
df_concatenado.to_csv('nuevo_archivo.txt', index=False, sep=',')
```

```python
# Guardar el nuevo dataframe en un archivo csv
df_concatenado.to_csv('nuevo_archivo.csv', index=False)
```

```python
optChainCalls = df.iloc[:, [2, 4, 7, 8, 13, 15, 17, 18, 19, 31]]
optChainCalls
```

```python
#optChainCalls[' [C_IV]'] = optChainCalls[' [C_IV]'].astype(float)
#optChainCalls[' [C_LAST]'] = optChainCalls[' [C_LAST]'].astype(float)
#optChainCalls[' [C_BID]'] = optChainCalls[' [C_BID]'].astype(float)
#optChainCalls[' [C_ASK]'] = optChainCalls[' [C_ASK]'].astype(float)

optChainCalls[' [QUOTE_DATE]'] = pd.to_datetime(optChainCalls[' [QUOTE_DATE]'])
```

```python
optChainCalls.info()
```

#### FILTRANDO DATA

```python
#dejando solo DTE entre 7 y 21
optChainCalls = optChainCalls[(optChainCalls[' [DTE]'] >= 7) & (optChainCalls[' [DTE]'] <= 21)]
optChainCalls
```

```python
#ahora seleccionaremos el delta más cercano a 0.5 (ATM) de cada fecha
idx_min_abs_diff = optChainCalls.groupby(' [QUOTE_DATE]')[' [C_DELTA]'].apply(lambda x: (x - 0.5).abs().idxmin())

optChainCalls = optChainCalls.loc[idx_min_abs_diff]

optChainCalls
```

```python
optChainCalls.info()
```

```python
#quedan algunos stype object, los transformaremos
optChainCalls[' [C_IV]'] = optChainCalls[' [C_IV]'].astype(float)
optChainCalls[' [C_LAST]'] = optChainCalls[' [C_LAST]'].astype(float)
optChainCalls[' [C_BID]'] = optChainCalls[' [C_BID]'].astype(float)
optChainCalls[' [C_ASK]'] = optChainCalls[' [C_ASK]'].astype(float)

optChainCalls.dtypes
```

```python
optChainCalls[' [C_PRIMA]'] = (optChainCalls[' [C_BID]']+optChainCalls[' [C_ASK]'])/2
optChainCalls.tail(2)
```


#### GRAFICAR

```python
fig, ax1 = plt.subplots(figsize=(12, 6))

#Graficar los valores de 'UNDERLYING_LAST' o precio de TSLA en el eje izquierdo
ax1.plot(optChainCalls[' [QUOTE_DATE]'], optChainCalls[' [UNDERLYING_LAST]'], color='blue', label='Cotización')
ax1.set_ylabel('UNDERLYING LAST', color='blue')
#ax1.tick_params('y', color='blue')
#ax1.legend(loc='lower left')

#Crear el eje derecho y graficar los valores de 'C_LAST' en el eje derecho
ax2 = ax1.twinx()
ax2.plot(optChainCalls[' [QUOTE_DATE]'], optChainCalls[' [C_PRIMA]'], color='red', label='Prima Calls ATM 7-21 DTE')
ax2.set_ylabel('PRIMA CALLS ATM', color='red')
#ax2.tick_params('y', colors='red')
#ax2.legend(loc='lower right')

# Configurar el eje x con las fechas
plt.xticks(rotation=90)
plt.xlabel(' [QUOTE_DATE]')

plt.title('ACCIÓN v/s PRIMA CALLS ATM de TSLA')

# Mostrar el gráfico
#plt.tight_layout()
plt.show()
```
