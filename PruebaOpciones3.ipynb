```python
df[' [QUOTE_DATE]'] = pd.to_datetime(df[' [QUOTE_DATE]'])
```

```python
optChainCalls = df.iloc[:, [2, 4, 7, 8, 13, 15, 17, 18, 19]]
optChainCalls
```

```python
optChainCalls.info()
```

```python
# Usar to_numeric() con la opción errors='coerce': La función pd.to_numeric() tiene una opción errors que te permite controlar cómo manejar los errores al convertir. Si usas errors='coerce', los valores no válidos se convertirán a NaN:

optChainCalls[' [C_DELTA]'] = pd.to_numeric(optChainCalls[' [C_DELTA]'], errors='coerce')
optChainCalls[' [C_DELTA]'] = optChainCalls[' [C_DELTA]'].astype(float)
```

```python
optChainCalls[' [C_IV]'] = pd.to_numeric(optChainCalls[' [C_IV]'], errors='coerce')
optChainCalls[' [C_LAST]'] = pd.to_numeric(optChainCalls[' [C_LAST]'], errors='coerce')

optChainCalls[' [C_IV]'] = optChainCalls[' [C_IV]'].astype(float)
optChainCalls[' [C_LAST]'] = optChainCalls[' [C_LAST]'].astype(float)
```

```python
optChainCalls[' [C_BID]'] = pd.to_numeric(optChainCalls[' [C_BID]'], errors='coerce')
optChainCalls[' [C_ASK]'] = pd.to_numeric(optChainCalls[' [C_ASK]'], errors='coerce')

optChainCalls[' [C_BID]'] = optChainCalls[' [C_BID]'].astype(float)
optChainCalls[' [C_ASK]'] = optChainCalls[' [C_ASK]'].astype(float)
```

```python
optChainCalls.info()
```

#### Filtrando datos

```python
# creando columna PRIMA como promedio de BID-ASK
optChainCalls[' [C_PRIMA]'] = (optChainCalls[' [C_BID]'] + optChainCalls[' [C_ASK]']) /2  
```

```python
# solo 7-14 DTE
optChainCalls = optChainCalls[(optChainCalls[' [DTE]'] >= 7) & (optChainCalls[' [DTE]'] <= 11)]
optChainCalls
```

```python
optChainCalls.info()
```

```python
# ahora seleccionaremos el delta más cercano a 0.5 (ATM) de cada fecha
idx_min_abs_diff = optChainCalls.groupby(' [QUOTE_DATE]')[' [C_DELTA]'].apply(lambda x: (x - 0.5).abs().idxmin())

optChainCalls = optChainCalls.loc[idx_min_abs_diff]

optChainCalls
```

Fucking split me cago la data, ante esto quedarme solo con fechas anterior al split.

```python
fechaSplit = pd.to_datetime('2022-08-25')
optChainCalls = optChainCalls.loc[optChainCalls[' [QUOTE_DATE]'] < fechaSplit]
```

```python
optChainCalls.tail(4)
```

Separar la data entre periodos pre-earnings

```python
earningsDateQ4_21 = pd.to_datetime('2022-01-26')
earningsDateQ1_22 = pd.to_datetime('2022-04-20')
earningsDateQ2_22 = pd.to_datetime('2022-07-20')
```

```python
# data periodos entre earnings
optChainCallsQ4_21 = optChainCalls.loc[optChainCalls[' [QUOTE_DATE]'] <= earningsDateQ4_21]
optChainCallsQ1_22 = optChainCalls.loc[(optChainCalls[' [QUOTE_DATE]'] > earningsDateQ4_21) & (optChainCalls[' [QUOTE_DATE]'] <= earningsDateQ1_22)]
optChainCallsQ2_22 = optChainCalls.loc[(optChainCalls[' [QUOTE_DATE]'] > earningsDateQ1_22) & (optChainCalls[' [QUOTE_DATE]'] <= earningsDateQ2_22)]
```
